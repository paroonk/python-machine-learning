```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy.io import loadmat
from scipy.optimize import minimize
from sklearn import preprocessing
#sns.set(context="talk")
plt.xkcd();
```

### Visualizing the dataset

```python
data = loadmat("ex5data1.mat")
X = data["X"]
print("X matrix shape =", X.shape)
y = data["y"]
print("y matrix shape =", y.shape)
Xval = data["Xval"]
print("Xval matrix shape =", Xval.shape)
yval = data["yval"]
print("yval matrix shape =", yval.shape)
Xtest = data["Xtest"]
print("Xtest matrix shape =", Xtest.shape)
ytest = data["ytest"]
print("ytest matrix shape =", ytest.shape)

plt.figure(figsize=(8, 6))
plt.scatter(X, y, c="r", marker="x")
plt.xlabel("Change in water level (x)")
plt.ylabel("Water flowing out of the dam (y)")
plt.title("Historical records on the change in the water level")
plt.show()
```

### Define hypothesis and cost function

```python
# Linear regression function
def hypo(theta, X):
    h = X.dot(theta)
    return h.reshape((-1, 1))

# Cost and gradient function
def cost(theta, X, y, lamb=0):
    m = len(y)
    h = hypo(theta, X)
    reg = lamb/(2*m) * np.sum(theta[1:]**2)
    cost = 1/(2*m) * np.sum((h-y)**2) + reg
    grad = 1/m * (h - y).T.dot(X).T
    reg_grad = lamb/m * theta[1:]
    grad[1:, 0] = grad[1:, 0] + reg_grad
    return cost, grad.ravel()
```

### Initialize parameters and training linear and polynomial regression at different polynomial degree

```python
degree = [1, 2, 8]
lamb = 0
cost_hist = np.zeros((X.shape[0]))
cost_val_hist = np.zeros((X.shape[0]))
plt.figure(figsize=(15, 8))
for i in range(len(degree)):
    
    # Polynomial Features
    poly = preprocessing.PolynomialFeatures(degree[i])
    X_poly = poly.fit_transform(X)
    Xval_poly = poly.fit_transform(Xval)
    Xtest_poly = poly.fit_transform(Xtest)
    # Standardization
    scaler = preprocessing.StandardScaler().fit(X_poly[:, 1:])
    X_poly[:, 1:] = scaler.transform(X_poly[:, 1:])
    Xval_poly[:, 1:] = scaler.transform(Xval_poly[:, 1:])
    Xtest_poly[:, 1:] = scaler.transform(Xtest_poly[:, 1:])
        
    # Initial theta
    theta = np.zeros([X_poly.shape[1]])
    
    for j in range(1, X.shape[0]+1):
        
        # Training Linear and Polynomial regression
        res = minimize(cost, theta, args=(X_poly[:j, :], y[:j], lamb), jac=True)
        cost_hist[j-1] = cost(res.x, X_poly[:j, :], y[:j], lamb)[0]
        cost_val_hist[j-1] = cost(res.x, Xval_poly, yval, lamb)[0]
        
    # Plot regression fit curve to data
    plt.subplot(2, 3, i+1)
    plt.scatter(X, y, c="r", marker="x")
    x_plot = np.array([np.linspace(X.min(), X.max(), 100)]).T
    x_plot_poly = poly.fit_transform(x_plot)
    x_plot_poly[:, 1:] = scaler.transform(x_plot_poly[:, 1:])
    y_plot = hypo(res.x, x_plot_poly).flatten()
    plt.plot(x_plot, y_plot, "b--")
    plt.title("Polynomial degree = {0}".format(degree[i]))
    plt.xlabel("Change in water level (x)")
    if i == 0: plt.ylabel("Water flowing out of the dam (y)")
    plt.text(-5, 34, "Cost_Train = {0:.2f}".format(cost(res.x, X_poly, y, lamb)[0]), ha="center")
    plt.text(-5, 30, "Cost_Val = {0:.2f}".format(cost(res.x, Xval_poly, yval, lamb)[0]), ha="center")
    plt.text(-5, 26, "Cost_Test = {0:.2f}".format(cost(res.x, Xtest_poly, ytest, lamb)[0]), ha="center")
    
    # Plot learning curve
    plt.subplot(2, 3, i+4)
    x_plot = range(1, len(cost_hist)+1)
    plt.plot(x_plot, cost_hist)
    plt.plot(x_plot, cost_val_hist)
    plt.xlabel("Number of training examples")
    plt.xticks(x_plot)
    if i == 0: plt.ylabel("Cost (J)")
    plt.legend(["Train", "Cross Validation"])

plt.suptitle("Polynomial fit and learning curve at different polynomial degree")
plt.subplots_adjust(top=0.88, hspace=0.35)
plt.show()
```

### Select regularized parameter (lambda) with cross validation set

```python
degree = 8
lamb = [0, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10]
cost_hist = np.zeros((len(lamb)))
cost_val_hist = np.zeros((len(lamb)))
plt.figure(figsize=(10, 6))
for i in range(len(lamb)):
    
    # Polynomial Features
    poly = preprocessing.PolynomialFeatures(degree)
    X_poly = poly.fit_transform(X)
    Xval_poly = poly.fit_transform(Xval)
    Xtest_poly = poly.fit_transform(Xtest)
    # Standardization
    scaler = preprocessing.StandardScaler().fit(X_poly[:, 1:])
    X_poly[:, 1:] = scaler.transform(X_poly[:, 1:])
    Xval_poly[:, 1:] = scaler.transform(Xval_poly[:, 1:])
    Xtest_poly[:, 1:] = scaler.transform(Xtest_poly[:, 1:])
    
    # Initial theta
    theta = np.zeros([X_poly.shape[1]])
    
    # Training Linear and Polynomial regression
    res = minimize(cost, theta, args=(X_poly, y, lamb[i]), jac=True)
    cost_hist[i] = cost(res.x, X_poly, y, lamb[i])[0]
    cost_val_hist[i] = cost(res.x, Xval_poly, yval, lamb[i])[0]

# Plot cross validation curve
plt.plot
plt.plot(lamb, cost_hist)
plt.plot(lamb, cost_val_hist)
plt.plot(lamb[np.argmin(cost_val_hist)], cost_val_hist[np.argmin(cost_val_hist)], "ro")
plt.annotate("Optimum lambda = {}".format(lamb[np.argmin(cost_val_hist)]), 
             xy=(lamb[np.argmin(cost_val_hist)], cost_val_hist[np.argmin(cost_val_hist)]), 
             xytext=(0.1, 30), arrowprops=dict(facecolor="r", shrink=0.05))
plt.xscale("log")
plt.xlabel("Regularized Parameter (Lambda)")
plt.ylabel("Cost (J)")
plt.legend(["Train", "Cross Validation"])
plt.title("Cross validation curve")
plt.show()

lamb = lamb[np.argmin(cost_val_hist)]
```
