```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy.optimize import minimize
#sns.set(context="talk")
plt.xkcd();
```

### Visualizing the dataset

```python
df = pd.read_csv("ex2data2.txt", names=["Test1", "Test2", "Accepted"])
X = df.values[:, 0:len(df.columns)-1]
y = df.values[:, len(df.columns)-1:len(df.columns)]

sns.lmplot(x="Test1", y="Test2", data=df,
           fit_reg=False, # No regression line, 
           hue="Accepted",
           size=6, # Height in inch
           aspect=1)  # Ratio of width to height
plt.show()

df.describe()
```

### Define hypothesis and cost function

```python
# Logistic regression function
def sigmoid(z):
    h = 1 / (1 + np.exp(-z))
    return h

# Cost and gradient function
def cost(theta, X, y, lamb=0):
    m = len(y)
    h = sigmoid(X.dot(theta).reshape(y.shape))
    reg = lamb/(2*m) * np.sum(theta[1:]**2)
    cost = 1/m * np.sum(-y*np.log(h) - (1-y)*np.log(1-h)) + reg
    grad = 1/m * (h - y).T.dot(X).T
    reg_grad = lamb/m * theta[1:]
    grad[1:, 0] = grad[1:, 0] + reg_grad
    return cost, grad.ravel()

# Map Feature
def mapFeature(X, degree=1):
    nrows = X.shape[0]
    ncols = int((degree+2)*(degree+1)/2)
    X_map = np.ones((nrows, ncols))
    for i in range(1, degree+1):
        for j in range(0, i+1):
            X_map[:, int((i+1)*i/2)+j] = (X[:, 0]**(i-j))*(X[:, 1]**j)
    return X_map
```

### Initialize parameters

```python
degree = 6
X = mapFeature(X, degree)
```

### Plot decision boundary at different regularization parameters (lambda)

```python
def predict(theta, X):
    return (sigmoid(X.dot(theta).reshape(y.shape))>=0.5)

lamb = [0.0001, 1, 100]
plt.figure(figsize=(15, 4))
for i in range(len(lamb)):
    theta = np.zeros([X.shape[1]])
    res = minimize(cost, theta, args=(X, y, lamb[i]), jac=True, method="TNC")
    plt.subplot(1, 3, i+1)
    plt.scatter(df.loc[df["Accepted"]==0].values[:, 0], df.loc[df["Accepted"]==0].values[:, 1], c="teal")
    plt.scatter(df.loc[df["Accepted"]==1].values[:, 0], df.loc[df["Accepted"]==1].values[:, 1], c="orange")
    x_plot = np.linspace(df.Test1.min(), df.Test1.max(), 100)
    y_plot = np.linspace(df.Test2.min(), df.Test2.max(), 100)
    z_plot = np.zeros((len(x_plot), len(y_plot)))
    for j in range(len(x_plot)):
        for k in range(len(y_plot)):
            x_predict = np.array([[x_plot[j], y_plot[k]]])
            x_predict = mapFeature(x_predict, degree)
            z_plot[j, k] = sigmoid(x_predict.dot(res.x)).ravel()[0]
    plt.contour(x_plot, y_plot, z_plot.T, [0.5], colors="red")
    acc = np.mean((predict(res.x, X)==y))
    plt.xlabel("Lambda = {0}\nAccuracy = {1:.2f}%".format(lamb[i], acc*100))
plt.suptitle("Decision boundary at different regularization parameters (lambda)")
plt.subplots_adjust(top=0.88)
plt.show()
```
